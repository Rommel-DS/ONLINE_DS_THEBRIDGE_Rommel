### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

#### 1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

```python
#sacar las mas pobladas
mas_pobladas = df_california.loc[df_california.population_total.nlargest(5).index, ["city","population_total"]]

#modificar los nombres y remplazarlos
modificadas = {"LosAngeles": "Los Angeles", "SanDiego": "San Diego", "SantaAna": "Santa Ana"}
ciudades = mas_pobladas["city"].replace (modificadas).to_list()

#crear la figura
plt.figure(figsize = (10,5))
plt.bar(x= ciudades, height= mas_pobladas["population_total"]/1000, color = ("blue","gray","red","olive","brown"), alpha = 0.5)
plt.ylim([0,5000])
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades")


mas_pobladas.set_index("city", inplace = True)
for indice, ciudad in enumerate(mas_pobladas.index):
    poblacion = mas_pobladas.loc[ciudad,"population_total"]/1000
    plt.text(ciudades[indice], poblacion + 50, f"{poblacion}", ha = "center")
    
plt.title("Ciudades más pobladas de California")
plt.show()
```

#### 2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)

```python
plt.figure (figsize = (10,5))
plt.hist(df_california["area_total_km2"], color = "steelblue", bins = 50, label= "Hist Area")
plt.legend()
plt.ylabel("Número de ciudades", fontdict=dict(family ="Bradley Hand itc"), size = 18)
plt.xlabel("Superficie($km^2$)", fontdict=dict(family="Bradley Hand itc"), size = 18)
plt.show()
```

#### 3. El scatter de área de agua de un ciudad con su poblacion.

```python
plt.figure (figsize=(10,5))
plt.scatter (df_california["area_water_km2"], df_california["population_total"]/1000, color="green", alpha = 0.5)
plt.ylabel("Población (miles de habitantes)")
plt.xlabel("Área de agua en la ciudad ($km^2$)")
plt.title("Población por kilómetro cuadrado de agua")


plt.show()
```

#### 4. El boxplot de la altura de las ciudades.

```python
plt.figure (figsize=(10,5))
plt.boxplot(df_california["elevation_m"],vert=False, patch_artist=True, boxprops=dict(facecolor="red"), whiskerprops=dict(color="black"), medianprops=dict(color="black"))
plt.xlabel("Altitud(m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California");
```

```python
# Juntar todas

fig, ax = plt.subplots (2,2, figsize = (20,20))

#primera grafica
mas_pobladas = df_california.loc[df_california.population_total.nlargest(5).index, ["city","population_total"]]
modificadas = {"LosAngeles": "Los Angeles", "SanDiego": "San Diego", "SantaAna": "Santa Ana"}
ciudades = mas_pobladas["city"].replace (modificadas).to_list()

ax[0,0].bar(x= ciudades, height= mas_pobladas["population_total"]/1000, color = ("blue","gray","red","olive","brown"), alpha = 0.5)
ax[0,0].set_ylim([0,5000])
ax[0,0].set_ylabel("Habitantes (miles)")
ax[0,0].set_xlabel("Ciudades")

mas_pobladas.set_index("city", inplace = True)
for indice, ciudad in enumerate(mas_pobladas.index):
    poblacion = mas_pobladas.loc[ciudad,"population_total"]/1000
    ax[0,0].text(ciudades[indice], poblacion + 50, f"{poblacion}", ha = "center")
    
ax[0,0].set_title("Ciudades más pobladas de California")

#segunda
ax[0,1].hist(df_california["area_total_km2"], color = "steelblue", bins = 50, label= "Hist Area", alpha = 0.7)
ax[0,1].legend()
ax[0,1].set_ylabel("Número de ciudades", fontdict=dict(family ="Bradley Hand itc"), size = 18)
ax[0,1].set_xlabel("Superficie($km^2$)", fontdict=dict(family="Bradley Hand itc"), size = 18)

#tercera
ax[1,0].scatter (df_california["area_water_km2"], df_california["population_total"]/1000, color="green", alpha = 0.5)
ax[1,0].set_ylabel("Población (miles de habitantes)")
ax[1,0].set_xlabel("Área de agua en la ciudad ($km^2$)")
ax[1,0].set_title("Población por kilómetro cuadrado de agua")

#cuarta
ax[1,1].boxplot(df_california["elevation_m"],vert=False, patch_artist=True, boxprops=dict(facecolor="red"), whiskerprops=dict(color="black"), medianprops=dict(color="black"))
ax[1,1].set_xlabel("Altitud(m)")
ax[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California");
```
